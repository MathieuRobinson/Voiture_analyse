```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(12,4),dpi=300)
sns.scatterplot(x='horsepower',y='acceleration',data=df_new,hue='origin',s=60,style='origin')
plt.title("Diagramme de dispersion de l'horsepower et de l'acceleration",fontweight='heavy',color='blue')
plt.show()
```

```python
cylindre_europe=new_region.loc['europe'][['cylinders','model_year']]
```

```python
cylindre_europe
```

```python
cylindre_japan=new_region.loc['japan'][['cylinders','model_year']]
```

```python

# Creation de la figure 
plt.figure(figsize=(9,5),dpi=150)
sns.set(style='darkgrid')
sns.barplot(x = "model_year", y = "cylinders", data = cylindre_europe, palette='plasma')
plt.xticks(rotation = 60)
plt.ylabel("Cylinders",color='Blue')
plt.xlabel("Year",color='Red')
plt.title('Modele par annee des cylinders (EUROPE)')
```

```python

# Creation de la figure 
plt.figure(figsize=(8,5),dpi=150)
sns.barplot(x = "model_year", y = "cylinders", data = cylindre_japan, palette='Dark2')
plt.xticks(rotation = 60)
plt.ylabel("Cylinders",color='Blue')
plt.xlabel(" Year",color='Red')
plt.title('Modele par annee des cylinders (JAPAN)')
```
